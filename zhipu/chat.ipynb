```python
import time

# 鉴权
import zhipuai

# 填写你的api key
# zhipuai.api_key="your api key"
```

```python
# 同步调用
res = zhipuai.model_api.invoke(
    model="chatglm_turbo",
    prompt=[
        {"role": "user", "content": "你好"},
        {"role": "assistant", "content": "我是人工智能助手"},
        {"role": "user", "content": "你叫什么名字"},
        {"role": "assistant", "content": "我叫chatGLM"},
        {"role": "user", "content": "你都可以做些什么事"},
    ],
    temperature=0.95,
    top_p=0.7,
    incremental=True
)
print(res)
```

```python
# 异步调用
import time

async_task = zhipuai.model_api.async_invoke(
    model="chatglm_turbo",
    prompt=[
        {"role": "user", "content": "你好"},
        {"role": "assistant", "content": "我是人工智能助手"},
        {"role": "user", "content": "你叫什么名字"},
        {"role": "assistant", "content": "我叫chatGLM"},
        {"role": "user", "content": "你都可以做些什么事"},
    ],
    temperature=0.95,
    top_p=0.7,
    incremental=True
)
print(async_task)

# 处理状态，PROCESSING（处理中），SUCCESS（成功），FAIL（失败）
while async_task["data"]["task_status"] in {"PROCESSING"}:
    async_task = zhipuai.model_api.query_async_invoke_result(async_task["data"]["task_id"])
    print(async_task)
    time.sleep(5)
```

```python
# SSE调用
response = zhipuai.model_api.sse_invoke(
    model="chatglm_turbo",
    #     本条信息作者的角色，可选择user 或 assistant
    #    - user 指用户角色输入的信息
    #   - assistant 指模型返回的信息
    prompt=[{"role": "user", "content": "人工智能"}],
    # 用温度取样的另一种方法，称为核取样
    # 取值范围是：(0.0, 1.0) 开区间，不能等于 0 或 1，默认值为 0.7
    # 模型考虑具有 top_p 概率质量tokens的结果
    # 例如：0.1 意味着模型解码器只考虑从前 10% 的概率的候选集中取tokens
    # 建议您根据应用场景调整 top_p 或 temperature 参数，但不要同时调整两个参数
    top_p=0.7,
    #  采样温度，控制输出的随机性，必须为正数
    # 取值范围是：(0.0,1.0]，不能等于 0,默认值为 0.95
    # 值越大，会使输出更随机，更具创造性；值越小，输出会更加稳定或确定
    # 建议您根据应用场景调整 top_p 或 temperature 参数，但不要同时调整两个参数
    temperature=0.9,
    # SSE接口调用时，用于控制每次返回内容方式是增量还是全量，不提供此参数时默认为增量返回
    # - true 为增量返回
    # - false 为全量返回
    incremental=True,
    # 用于控制每次返回内容的类型，空或者没有此字段时默认按照json_string返回
    # - json_string 返回标准的 JSON 字符串
    # - text 返回原始的文本内容
    return_type="text"
)

for event in response.events():
    if event.event == "add":
        print(event.data)
    elif event.event == "error" or event.event == "interrupted":
        print(event.data)
    elif event.event == "finish":
        print(event.data)
        print(event.meta)
    else:
        print(event.data)
```
